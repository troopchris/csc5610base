```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from decision_boundaries import linear_decision_boundary_classifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
setosa_df = pd.read_csv("setosa_data.csv")

setosa_df.head()
```

```python
sns.scatterplot(x="sepal_length (cm)", y="sepal_width (cm)", hue="label", data= setosa_df)
plt.plot([4, 8], [0, 8], 'black', label = "y=2x-8")
plt.plot([4, 8], [2.1, 5.1], 'red', label = "y=0.75x-0.9")
plt.plot([4, 8], [4, -4], 'purple', label = "y=-2x+12")
plt.legend()
plt.rcParams['savefig.dpi'] = 300
plt.show()
```

standard form 

-2x+y+8=0
-0.75+y+0.9=0
2x+y-12=0

```python
dec_bound_vec = np.array([-2, 1.00, 8])
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(dec_bound_vec, features, class_names=class_names)
print(pred_labels)

print (accuracy_score(setosa_df['label'], pred_labels))
```

```python
dec_bound_vec = np.array([-0.75, 1.00, 0.9])
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(dec_bound_vec, features, class_names=class_names)
print(pred_labels)

print (accuracy_score(setosa_df['label'], pred_labels))
```

```python
dec_bound_vec = np.array([2, 1.00, -12])
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
class_names = ["not setosa", "setosa"]
pred_labels = linear_decision_boundary_classifier(dec_bound_vec, features, class_names=class_names)
print(pred_labels)

print (accuracy_score(setosa_df['label'], pred_labels))
```

```python
setosa_training_df = pd.read_csv("setosa_training.csv")
training = setosa_training_df[["sepal_length (cm)", "sepal_width (cm)"]].values
trainingLabel = setosa_training_df[["label"]].values

trainingLabel = np.ravel(trainingLabel)

le.fit(trainingLabel)

print(le.transform(trainingLabel))
```

```python
setosa_testing_df = pd.read_csv("setosa_testing.csv")
testing= setosa_testing_df[["sepal_length (cm)", "sepal_width (cm)"]].values
testingLabel = setosa_testing_df[["label"]].values

testingLabel = np.ravel(testingLabel)

le.fit(testingLabel)

print(le.transform(testingLabel))
```

```python
setosa_target_df = pd.read_csv("setosa_target.csv")
target = setosa_target_df[["sepal_length (cm)", "sepal_width (cm)"]].values
targetLabel = setosa_target_df[["label"]].values

targetLabel = np.ravel(targetLabel)

le.fit(targetLabel)

print(le.transform(targetLabel))
```

```python
model = LogisticRegression(random_state=0)
model.fit(training, trainingLabel)

print(model.coef_)
print(model.intercept_)
```

3.16919523x-2.57279307y-9.15324246=0
y=1.2318x-3.5577

```python
train_predictions = model.predict(training)
train_accuracy = accuracy_score(trainingLabel, train_predictions)
print(train_accuracy)
```

```python
test_predictions = model.predict(testing)
test_accuracy = accuracy_score(testingLabel, test_predictions)
print(test_accuracy)
```

```python
target_predictions = model.predict(target)
target_accuracy = accuracy_score(targetLabel, target_predictions)
print(target_accuracy)
```
