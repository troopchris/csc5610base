```python
sns.boxplot(x=Awards_Batting_HoF_df["HallOfFame_df_inducted"], y=Awards_Batting_HoF_df["BattingPost_df_GIDP"])

samples_by_group = []
for value in set(Awards_Batting_HoF_df["HallOfFame_df_inducted"]):
    mask = Awards_Batting_HoF_df["HallOfFame_df_inducted"] == value
    samples_by_group.append(Awards_Batting_HoF_df["BattingPost_df_GIDP"][mask])
    
stat, p = stats.kruskal(*samples_by_group)
print(p)
```
