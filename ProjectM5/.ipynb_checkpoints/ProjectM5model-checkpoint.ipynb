```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow
from scipy import stats
import datetime as dt
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
flights_df = pd.read_feather('allflights_df.feather')
```

```python
flights_df.info()
```

pull out date, flight count, cancelled count, completed count and holiday for each day

```python
flightsDay_df = flights_df[['date' , 'flight_count', 'cancelled_count', 'completed_count', 'Holiday']]
```

```python
flightsDay_df = flightsDay_df.drop_duplicates(subset='date', keep='last')
flightsDay_df.head()
```

```python
flightsDay_df['Holiday'] = flightsDay_df['Holiday'].astype(bool)
```

```python
plt.plot(flightsDay_df['date'], flightsDay_df['flight_count'], label='data', color='black')
plt.xticks(rotation=45)
plt.xlabel("date", fontsize=16)
plt.ylabel("flight_count", fontsize=16)
plt.show()
```

```python
train_end_date = '2017-01-01'
train_df = flightsDay_df[flightsDay_df['date'] < train_end_date]
test_df = flightsDay_df[flightsDay_df['date'] >= train_end_date]

plt.figure(figsize=(14, 7))
plt.plot(train_df['date'], train_df['flight_count'], label='Training Data', color='black')
plt.plot(test_df['date'], test_df['flight_count'], label='Testing Data', color='grey')
plt.title('Time Series Data with Training and Testing Sets')
plt.xlabel('Date')
plt.ylabel('Flight Count')
plt.legend()
plt.show()

train_df['elapsed_days'] = (train_df['date'] - train_df['date'].min()).dt.days

X_train = train_df[['elapsed_days']]
y_train = train_df[['flight_count']]

model = LinearRegression()
model.fit(X_train, y_train)

slope = model.coef_[0]
intercept = model.intercept_

slope, intercept, r_value, p_value, std_err = stats.linregress(X_train.values.flatten(), y_train.values.flatten())

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"P-value: {p_value}")

plt.figure(figsize=(14, 7))
plt.plot(train_df['date'], train_df['flight_count'], label='Training Data', color='black')
plt.plot(test_df['date'], test_df['flight_count'], label='Testing Data', color='grey')

regression_line = intercept + slope * train_df['elapsed_days']
plt.plot(train_df['date'], regression_line, label='Regression Line', color='red')

plt.title('Time Series Data with Linear Regression Line')
plt.xlabel('Date')
plt.ylabel('Flight Count')
plt.legend()
plt.show()
```

```python
flightsDay_df['elapsed_days'] = (flightsDay_df['date'] - flightsDay_df['date'].min()).dt.days

train_end_date = '2017-01-01'

training_df = flightsDay_df[flightsDay_df['date'] < train_end_date]
testing_df = flightsDay_df[flightsDay_df['date'] >= train_end_date]

print(len(training_df), len(testing_df))
```

```python
train_X = training_df[["elapsed_days"]].values
test_X = testing_df[["elapsed_days"]].values

train_y = training_df["flight_count"].values
testing_y = testing_df["flight_count"].values

ridge_model = Ridge()

ridge_model.fit(train_X, train_y)

testing_df["ridge_pred"] = ridge_model.predict(test_X)

training_df["ridge_pred"] = ridge_model.predict(train_X)
```

```python
fig = plt.gcf()
fig.set_dpi(300)

ax = sns.lineplot(data=training_df, x="date", y="flight_count", label="Training", color="black")
ax = sns.lineplot(data=testing_df, x="date", y="flight_count", label="Testing (True)", color="gray")
ax = sns.lineplot(data=testing_df, x="date", y="ridge_pred", color="tab:orange")
ax = sns.lineplot(data=training_df, x="date", y="ridge_pred", label="Predicted", color="tab:orange")
plt.xticks(rotation=45)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Flight Count", fontsize=16)
plt.legend(loc="upper left", fontsize=12)
```

```python
ridge_rmse = np.sqrt(mean_squared_error(testing_y, testing_df["ridge_pred"]))
ridge_mape = ridge_rmse / np.mean(testing_df["flight_count"])

print("RMSE: {:.0f}".format(ridge_rmse))
print("MAPE: {:.0%}".format(ridge_mape))
```

```python
first_date = training_df["date"][841]

flightsDay_df["elapsed_days"] = flightsDay_df["date"].apply(lambda d: (d - first_date).days)
flightsDay_df["month"] = flightsDay_df["date"].apply(lambda d: d.month)
flightsDay_df["day_of_week"] = flightsDay_df["date"].apply(lambda d: d.weekday())

dummies_month = pd.get_dummies(flightsDay_df["month"], prefix="month")
dummies_weekday = pd.get_dummies(flightsDay_df["day_of_week"], prefix="day_of_week")

merged_with_dummies = dummies_month.merge(flightsDay_df, left_index=True, right_index=True)
merged_with_dummies = merged_with_dummies.merge(dummies_weekday, left_index=True, right_index=True)

testing_cutoff = int(len(flightsDay_df) * 0.8)

training_df = merged_with_dummies.iloc[:testing_cutoff].copy()
testing_df = merged_with_dummies.iloc[testing_cutoff:].copy()

dummies_weekday.head()
```

```python
features = ["month_{}".format(m) for m in range(1, 13)]
features.append("elapsed_days")

train_X = training_df[features].values

print(train_X.shape)

test_X = testing_df[features].values

train_y = training_df["flight_count"].values
testing_y = testing_df["flight_count"].values

ridge_model = Ridge()

ridge_model.fit(train_X, train_y)

testing_df["ridge_pred"] = ridge_model.predict(test_X)

training_df["ridge_pred"] = ridge_model.predict(train_X)
```

```python
fig = plt.gcf()
fig.set_dpi(150)

ax = sns.lineplot(data=training_df, x="date", y="flight_count", label="Training", color="black")
ax = sns.lineplot(data=testing_df, x="date", y="flight_count", label="Testing (True)", color="gray")
ax = sns.lineplot(data=testing_df, x="date", y="ridge_pred", label="Predicted", color="tab:orange")
ax = sns.lineplot(data=training_df, x="date", y="ridge_pred", color="tab:orange")
plt.xticks(rotation=45)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Count", fontsize=16)
plt.legend(loc="upper left", fontsize=12)
```

```python
ridge_rmse = np.sqrt(mean_squared_error(testing_y, testing_df["ridge_pred"]))
ridge_mape = ridge_rmse / np.mean(testing_df["flight_count"])

print("Ridge RMSE: {:.0f}".format(ridge_rmse))
print("Ridge MAPE: {:.0%}".format(ridge_mape))
```
