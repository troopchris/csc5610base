```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow
from scipy import stats
import datetime as dt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
flights_df = pd.read_feather('allflights_df.feather')
```

```python
flightsDay_df = flights_df[['date' , 'flight_count', 'cancelled_count', 'completed_count', 'Holiday']]
```

```python
flightsDay_df = flightDay_df.drop_duplicates(subset='date', keep='last')
flightsDay_df.head()
```

```python
fig = plt.gcf()
fig.set_dpi(300)

ax = sns.lineplot(data=flightsDay_df, x="date", y="flight_count", color="k")
plt.xticks(rotation=45)
plt.xlabel("date", fontsize=16)
plt.ylabel("flight_count", fontsize=16)
```

```python
testing_cutoff = int(len(flightsDay_df) * 0.8)

training_df = flightsDay_df.iloc[:testing_cutoff].copy()
testing_df = flightsDay_df.iloc[testing_cutoff:].copy()

print(len(training_df), len(testing_df))
```

```python
fig = plt.gcf()
fig.set_dpi(300)

ax = sns.lineplot(data=training_df, x="date", y="flight_count", label="Training", color="black")
ax = sns.lineplot(data=testing_df, x="date", y="flight_count", label="Testing", color="gray")
plt.xticks(rotation=45)
plt.xlabel("Date", fontsize=16)
plt.ylabel("Flight Count", fontsize=16)
plt.legend(fontsize=14)
```

```python
training_df['elapsed_days'] = (training_df['date'] - training_df['date'].min()).dt.days
```

```python
x = training_df['elapsed_days']
y = training_df['flight_count']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant relationship between elapsed time and flight count.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between elapsed time and flight count.")
```

```python
plt.figure(figsize=(10, 6))
ax = sns.lineplot(data=training_df, x="date", y="flight_count", label="Training", color="black")
ax = sns.lineplot(data=testing_df, x="date", y="flight_count", label="Testing", color="gray")
plt.plot(training_df['date'], intercept + slope * training_df['elapsed_days'], color='red', label='Regression Line')
plt.title('Flight Count Over Time with Linear Regression')
plt.xlabel('Date')
plt.ylabel('Flight Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
features = ["month_{}".format(m) for m in range(1, 13)]
features.append("elapsed_days")

train_X = training_df[features].values

print(train_X.shape)

test_X = testing_df[features].values

train_y = training_df["flight_count"].values
testing_y = testing_df["flight_count"].values

lr_reg_model = LinearRegression()

lr_reg_model.fit(train_X, train_y)

testing_df["lr_pred"] = lr_reg_model.predict(test_X)

training_df["lr_pred"] = lr_reg_model.predict(train_X)
```
