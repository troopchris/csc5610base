replacing all -1 cylinders with pd.NA
replacing all 3 year guarantee periods with 36 months

```python
carPrice_df = carPrice_df.replace({"Cylinders": -1}, pd.NA)
carPrice_df = carPrice_df.replace({"Guarantee_Period": 3}, 36)
```

```python
carPrice_df.describe()
```

```python
fig, ax = plt.subplots()

ax.scatter(carPrice_df["Age_08_04"], carPrice_df["Price"])
ax.set_xlabel("Age of Car in Months")
ax.set_ylabel("Price of Car")
ax.set_title("Age vs Price")

plt.rcParams['savefig.dpi'] = 300
plt.savefig('priceAgeScatter.png') 
plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(carPrice_df["KM"], carPrice_df["Price"])
ax.set_xlabel("KM")
ax.set_ylabel("Price of the Car")
ax.set_title("KM vs Price")

plt.rcParams['savefig.dpi'] = 300
plt.savefig('priceKMScatter.png') 
plt.show()
```
