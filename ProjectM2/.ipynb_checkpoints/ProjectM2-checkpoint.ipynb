```python
cleanCarPrice_df = pd.read_feather("cleanCarPrice_df.feather")  
cleanCarPrice_df.astype({'Id':'int32', 'Price':'float64', 'Fuel_Type': 'category', 'Color':'category', 'Automatic':'category','CC' : 'category','HP' : 'category' ,'Doors':'category', 'Cylinders':'category', 'Gears':'category', 'Quarterly_Tax':'category', 'Mfr_Guarantee':'category', 'BOVAG_Guarantee':'category', 'Guarantee_Period':'category', 'ABS':'category', 'Airbag_1':'category',  'Airbag_2':'category', 'Airco':'category', 'Automatic_airco' : 'category' ,'Boardcomputer':'category', 'CD_Player':'category', 'Central_Lock':'category', 'Powered_Windows':'category', 'Power_Steering':'category', 'Radio':'category',  'Mistlamps':'category', 'Sport_Model':'category',  'Backseat_Divider':'category', 'Metallic_Rim':'category', 'Radio_cassette':'category', 'Parking_Assistant':'category',  'Tow_Bar':'category', 'Mfg_Date':'datetime64[ns]'}).dtypes
```

Age vs Price
KM vs Price
Weight Vs Price

```python
fig, ax = plt.subplots()

ax.scatter(cleanCarPrice_df["Age_08_04"], cleanCarPrice_df["Price"])
ax.set_xlabel("Age")
ax.set_ylabel("Price")
ax.set_title("Age vs Price")

plt.show()

res = stats.linregress(cleanCarPrice_df["Age_08_04"], cleanCarPrice_df["Price"])
slope, intercept, r, p, stderr = stats.linregress(cleanCarPrice_df["Age_08_04"], cleanCarPrice_df["Price"])

plt.plot(cleanCarPrice_df["Age_08_04"], cleanCarPrice_df["Price"], 'o', label='original data')
plt.plot(cleanCarPrice_df["Age_08_04"], res.intercept + res.slope*cleanCarPrice_df["Age_08_04"], 'r', label='fitted line')
plt.legend()
plt.show()

print(f"slope: {slope} intercept: {intercept} r: {r} p: {p} stderr: {stderr}")
```

linear regression model

```python
y, X = dmatrices("Price ~ Age_08_04 + KM + HP + Color + CC + Doors + Gears + Quarterly_Tax + Weight + Mfr_Guarantee + BOVAG_Guarantee + Guarantee_Period + ABS + Airbag_1 + Airbag_2 + Airco + Automatic_airco + Boardcomputer + CD_Player + Central_Lock + Powered_Windows + Power_Steering + Mistlamps + Backseat_Divider + Metallic_Rim + Parking_Assistant + Tow_Bar",
                 data=cleanCarPrice_df, return_type="dataframe")
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

```python
cleanCarPrice_df["KM"] = cleanCarPrice_df["KM"] * 0.0001
cleanCarPrice_df["CC"] = cleanCarPrice_df["CC"] * 0.1
cleanCarPrice_df["Weight"] = cleanCarPrice_df["Weight"] * 0.1
```

```python
y, X = dmatrices("Price ~ Age_08_04 + KM + HP + Color + CC + Doors + Quarterly_Tax + Weight + Mfr_Guarantee + BOVAG_Guarantee + Guarantee_Period + ABS + Airbag_1 + Airbag_2 + Airco + Automatic_airco + Boardcomputer + CD_Player + Central_Lock + Powered_Windows +  Mistlamps + Backseat_Divider + Metallic_Rim + Tow_Bar",
                 data=cleanCarPrice_df, return_type="dataframe")
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

```python
pred_y = results.predict(X)
sns.scatterplot(x=y["Price"], y=pred_y)
plt.plot([0, 31000], [0, 31000], color="tab:orange", label='fitted line')
plt.xlabel("True Value", fontsize=16)
plt.ylabel("Predicted Value", fontsize=16)
plt.legend()
plt.savefig('linearRegModel.png') 
```

```python
mse = mean_squared_error(y["Price"], pred_y)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y["Price"], pred_y)

print("RMSE : {:.1f}".format(rmse))
print("MAPE: {:.1%}".format(mape))
```

```python
sns.histplot(x=results.resid)
plt.xlabel("Residual", fontsize=16)
plt.ylabel("Count (Samples)", fontsize=16)
plt.savefig('residuals.png') 
```
